<b>Handiling null values</b>

**name and host_name**-- we are replacing missing values in feature name and host_name with 'Not Known'

**reviews_per_month** -- We are replacing missing values in feature reviews_per_month with the mean of the feature because it is continues

**Last_review**
From Last review we extracted the feature Quarter where we got the month from the feature and replaced the null values with mode of feature Quarter and labels the months 1-3 as Q1, 4-6 as Q2, 7-9 as Q3 and 10-12 as Q4.


```python
# name and host_name -we are replacing missing values in feature name and host_name with 'Not Known'
df['name'].fillna('Not Known',inplace=True)

df['host_name'].fillna('Not Known',inplace=True)
df.isnull().sum()
```

```python
# reviews_per_month -- We are replacing missing values in feature reviews_per_month with the mean of the feature because it is continues
df['reviews_per_month'].fillna(df['reviews_per_month'].mean(),inplace=True)
df.isnull().sum()
```

```python

df['last_review']=pd.to_datetime(df['last_review'])#converting the last_review to datetime
df['Quarter']=df['last_review'].dt.month# extracting month and assigning the series to Quarter
df['Quarter'].value_counts()
```

```python
df['Quarter'].fillna(df['Quarter'].mode()[0],inplace=True)#replacing the null values in Quarter with mode()
```

```python
Quarter={1:'Q1',2:'Q1',3:'Q1',4:'Q2',5:'Q2',6:'Q2',7:'Q3',8:'Q3',9:'Q3',10:'Q4',11:'Q4',12:'Q4'}
#mapping the each month to the Quarter values
df['Quarter']=df['Quarter'].map(Quarter)
```

```python
# For the featur minimum_nights the max values is 1250, which is more than 3 years which can't be true to we are making the values which are greater than 365 then replace it with mean.
df['minimum_nights']=df["minimum_nights"].apply(lambda x:round(df['minimum_nights'].mean()) if x>365 else x )
```

```python
# For the feature Price we can see the minimum price is 0.0. which can't be true so we can replace them with mean of the feature
df['price']=df['price'].apply(lambda x:df['price'].mean() if x==0 else x)

```

```python
# Droping unwanted columns
columns=['id','last_review']
df.drop(columns, axis=1,inplace=True)
df.head()
```
